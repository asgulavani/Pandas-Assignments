```python
##14. ** How many people have the word Chief in their job title? (This is pretty tricky) **
```

```python
sal[sal['JobTitle'].apply(lambda x:'CHIEF' in x.upper())]
```

```python
def find_chief(job):
    if 'CHIEF' in job.upper():
        return True
    else:
        return False
```

```python
sal[sal['JobTitle'].apply(lambda x:find_chief(x)) == True]
```

```python
sum(sal['JobTitle'].apply(lambda x:find_chief(x)))
```
